```python
data = gpd.read_file('il.shp')
```

```python
data.columns
```

```python
data[data['incorpname'] == 'Chicago'].head()
```

```python
data = data[data['incorpname'] == 'Chicago'].copy()
```

```python
data.head()
```

```python
len(data)
```

```python
data['tract'] = data['geoid'].astype(str)
```

```python
data['tract'] = data['tract'].str[5:11]
```

```python
data['tract'].value_counts()
```

```python
data2 = data.groupby('tract').mean(numeric_only=True)
```

```python
data2.reset_index(inplace=True)
```

```python
comm_areas = pd.read_csv('chicago_tract_comm_areas.csv')
```

```python
comm_areas.columns = [c.lower() for c in comm_areas.columns]
```

```python
comm_areas.head()
```

```python
comm_areas['tract'] = comm_areas['tract'].astype(str)
```

```python
for i, row in comm_areas.iterrows():
    tract = row['tract']
    if len(tract) == 5:
        comm_areas.at[i, 'tract'] = '0' + tract
```

```python
data2.head()
```

```python
comm_areas.head()
```

```python
data_merged = pd.merge(data2, comm_areas, on=['tract'])
```

```python
data_merged
```

```python
comm_data = data_merged.groupby('community area name').mean(numeric_only=True).reset_index()
```

```python
comm_data.head()
```

```python
comm_data.columns
```

```python
comm_data['tc_goal']
```

```python
cols_to_drop = ['geoid', 'treecanopy', 'area', 'popadjust', 'tc_goal', 'priority', 'tesctyscor', 'sumlev',
       'county', 'cousub', 'place', 'geoid2', 'chgoca', 'notes' ]
```

```python
comm_data.drop(columns=cols_to_drop, inplace=True)
```

```python
comm_data.to_csv('chicago_tree_equity.csv')
```
